```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

The following function is $$ f_1 = x^2 + y^2 $$ which has a local minimum at $$ (x,y) = (0,0) $$

So this will be our f1 function and (0,0,0) will be p1

I will define the function below and then graph it using (A) 3D plots, like [here](https://colab.research.google.com/drive/1btvbObh-nZ4MSC7QkjpS3RGpefN_msth), and then (B) density plots and then (C) contour plots, like [here](https://colab.research.google.com/drive/1pcvtvK6jITbp1Sf2nD2uEaDGpwUOA3IL)

Below is code for $f_1$ in a contour plot

```
def f1(x,y):
  return x*x + y*y

xline = np.linspace(-10,10,1000)
yline = np.linspace(-10,10,1000)
zline = f1(xline, yline)

X, Y = np.meshgrid(xline,yline)
Z = f1(X,Y)
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, 30, colors='black')
#Mark the local minimum
plt.scatter(0,0)
plt.show()
```

Below is $f_1$ in a density plot

```
plt.contourf(X,Y,Z, 40, cmap='RdGy')
plt.colorbar()
plt.scatter(0,0)
plt.show()
```

Below is $f_1$ in a 3D plot


```
fig = plt.figure()

ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 50)
ax.scatter(0,0,0, c='Red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(50, 10)
```

*Now I will define $f_2$ and plot it accordingly*

$f_2$ will be defined as 

$$ f_2 = -\sqrt{x} - \sqrt{y} $$

Which appears as a cup shape in the positive X, positive Y quadrant

Our maximum will be at $$ (x,y) = (0,0) $$

The following is $f_2$ in a contour plot

```
def f2(x,y):
  return -np.sqrt(x) - np.sqrt(y)

xline = np.linspace(0,10,1000)
yline = np.linspace(0,10,1000)
zline = f2(xline, yline)

X, Y = np.meshgrid(xline,yline)
Z = f2(X,Y)
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, 30, colors='black')
#Mark the local maximum
plt.scatter(0,0)
plt.show()
```

The following is $f_2$ in a density plot

```
plt.contourf(X,Y,Z, 40, cmap='RdGy')
plt.colorbar()
plt.scatter(0,0)
plt.show()
```

The following is $f_2$ in a 3D plot

```
fig = plt.figure()

ax = plt.axes(projection='3d')

ax.plot_surface(X,Y,Z)
ax.scatter(0,0,0, c='Red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```

Now I will define a final function, $f_3$

$$ f_3 = y * sin(x) $$

$$ \nabla f_3(0,0) = ( 0 * cos(0), sin(0) ) = (0,0) $$

So the gradient at $ (0,0) $ is the zero vector.

First I will define the function and plot it in a contour plot.

```
def f3(x,y):
  return y * np.sin(x)

xline = np.linspace(-10,10,1000)
yline = np.linspace(-10,10,1000)
zline = f3(xline, yline)

X, Y = np.meshgrid(xline,yline)
Z = f3(X,Y)
plt.style.use('seaborn-white')
plt.contour(X,Y,Z, 30, colors='black')
#Mark the local minimum
plt.scatter(0,0)
plt.show()
```

The following is $ f_3 $ in a density plot

```
plt.contourf(X,Y,Z, 40, cmap='RdGy')
plt.colorbar()
plt.scatter(0,0)
plt.show()
```

The following is $f_3$ in a 3D plot

```
fig = plt.figure()

ax = plt.axes(projection='3d')
ax.scatter(0,0,0, c='Red', marker = 'o')
ax.plot_surface(X,Y,Z)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.view_init(30, 10)
```
